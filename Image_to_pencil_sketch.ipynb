```python
import cv2
from google.colab.patches import cv2_imshow
from PIL import Image
from IPython.display import display
```

```python
#load image
org_img = cv2.imread("/content/cat.jpg")
```

```python
#Convert image to grayscale
gray_image = cv2.cvtColor(org_img, cv2.COLOR_BGR2GRAY)
```

```python
display(Image.fromarray(org_img))
```

```python
#Convert the Original image into Grayscale

gray_image=cv2.cvtColor(org_img,cv2.COLOR_BGR2GRAY)
display(Image.fromarray(gray_image))
```

```python
# Invert the Grayscale image

inv_gray_image=255-gray_image
display(Image.fromarray(inv_gray_image))
```

```python
blur_image=cv2.GaussianBlur(inv_gray_image,(35,35),0)
```

```python
display(Image.fromarray(blur_image))
```

```python
#Invert the Blurred Image

inverted_blur_image=255-blur_image
```

```python
# Creat the pencil Sketch Image

pencil_sketch=cv2.divide(gray_image,inverted_blur_image, scale=256.0)
```

```python
cv2_imshow(pencil_sketch)
```

```python
display(Image.fromarray(org_img))
```

```python
display(Image.fromarray(pencil_sketch))
```
